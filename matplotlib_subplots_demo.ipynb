```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()
ax.plot(np.random.rand(20), label='A', c='g', marker='o', markersize=5, linewidth=3)
ax.plot(np.random.rand(20), label='B', c='Pink', marker='o', markersize=5, linewidth=3)
ax.plot(np.random.rand(20), label='C', c='purple', marker='o', markersize=5, linewidth=3)
ax.set_title('First Graph')
plt.legend(loc=2)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

n_rows = 2
n_cols = 2
fig, axes = plt.subplots(n_rows, n_cols)

labels = ['A', 'B', 'C', 'D']
colors = ['red', 'green', 'blue', 'black']

for row_num in range(n_rows):
    for col_num in range(n_cols):
        ax = axes[row_num][col_num]
        idx = row_num * n_cols + col_num
        ax.plot(np.random.rand(20), label=labels[idx], color=colors[idx])
        ax.set_title(f'Plot ({row_num}, {col_num})')

        ax.legend()

fig.suptitle('Main Title')
fig.tight_layout()
plt.show()
```
